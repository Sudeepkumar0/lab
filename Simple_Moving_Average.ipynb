```python
# from cProfile import label

import pandas as pd
import matplotlib.pyplot as plt

path = "/content/daily-min-temperatures.csv"
df = pd.read_csv(path, parse_dates=['Date'], index_col='Date')


series = df['Temp']
window_size = 7
moving_avg = series.rolling(window=window_size).mean()
forecast = moving_avg.iloc[-1]

print(f"Forecast next value {window_size}-day moving average : {forecast:.3f}")

plt.figure(figsize=(12,6))
plt.plot(series,label='daoly min temp')
plt.plot(moving_avg,label=f"{window_size}-day moving average")
plt.xlabel='Date'
plt.ylabel('temp')
plt.title("SMA")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
